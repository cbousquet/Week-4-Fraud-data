```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Chargement des données

```python
fraud = pd.read_csv("Fraud_Data.csv")
```

```python
country = pd.read_csv("IpAddress_to_Country.csv")
```

```python
fraud.head()
```

```python
# On crée une colonne supplémentaire dans laquelle on indique le pays en fonction de l'adresse IP
fraud["country"] = None
```

```python
lower = country["lower_bound_ip_address"]
upper = country["upper_bound_ip_address"]
```

```python
idx = pd.IntervalIndex.from_arrays(country["lower_bound_ip_address"], country["upper_bound_ip_address"], closed='both')
ct = country.iloc[idx.get_indexer(fraud['ip_address'])]['country']
```

```python
fraud['country'] = ct.values
```

```python
fig = plt.figure(figsize=(25,4))
sns.barplot(x='country', y='class', data=fraud)
plt.xticks(rotation=90)
```
